```python
raw_navigation_data = navigation_events.load_events_data(os.path.join(os.pardir, "data", "navigation_events.csv"))
cleaned_navigation_data = navigation_events.preprocess_events_data(raw_navigation_data)

l = locals()
for col in cleaned_navigation_data.columns:
    l['ch_' + str(col)] = alt.Chart(cleaned_navigation_data).mark_bar().encode(
        alt.X("count()"),
        alt.Y(col),
        alt.Color(col)
)
```
